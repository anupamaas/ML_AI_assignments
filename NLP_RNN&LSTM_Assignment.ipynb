Importing necessary libraries

```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import gensim
from keras.preprocessing import text
from keras.utils import pad_sequences
```


To Load the dataset in Python environment

```python
data=pd.read_csv('/content/judge-1377884607_tweet_product_company.csv', encoding="ISO-8859-1")
```

```python
data
```

Dropping unnecessary columns

```python
data.drop('emotion_in_tweet_is_directed_at',axis=1,inplace=True)
```

Displaying the dataset with maximum column width

```python
pd.set_option('display.max_colwidth',None)
data.head()
```

Preprocessing of text data

```python
nltk.download('stopwords')
```

```python
stopword=nltk.corpus.stopwords.words('english')
```

```python
def sw_removed(text):
  sw =[i for i in text if i not in stopword]
  return sw
```

```python
data['tweet_text'] = data['tweet_text'].astype(str)
data['tweet_clear'] = data['tweet_text'].apply(lambda x: gensim.utils.simple_preprocess(x))
```

```python
data['tweet_clean']=data['tweet_clear'].apply(lambda x: sw_removed(x))
```

```python
data.head()
```

converting text to sequences and padding to get equal length for datas

```python
tokenizer = text.Tokenizer()
tokenizer.fit_on_texts(data['tweet_clean'])
tokenized_texts = tokenizer.texts_to_sequences(data['tweet_clean'])
X = pad_sequences(tokenized_texts, maxlen=50)
```

```python
X[7]
```

```python
vocabulary_size=len(tokenizer.word_index)+1
```

encoding y datas using one hot encoding

```python

y = pd.get_dummies(data['is_there_an_emotion_directed_at_a_brand_or_product'])

```

```python
y
```

Splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
seq_len=X.shape[1]
vocabulary_size=len(tokenizer.word_index)+1
```

```python
from keras.utils import to_categorical
train_targets = to_categorical(X, num_classes=vocabulary_size)
```

```python
from keras.layers import Dense, LSTM, Embedding
from keras.models import Sequential
```


#Adding  Embedding,LSTM & Dense Layers

```python
model = Sequential()
model.add(Embedding(vocabulary_size, 5, input_length=seq_len))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(60))
model.add(Dense( 60,activation = 'relu'))
model.add(Dense(4, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

Fitting the model

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.1)
```

```python
model.summary()
```

```python
y_pred=model.predict(X_test)
```
